## Realizando Imports

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import yfinance as yf
import lxml
```

## Lendo dados financeiros das empresas

```python
url = 'https://fundamentus.com.br/resultado.php'
agent = {"User-Agent": "Mozzila/5.0"}
response = requests.get(url, headers=agent)
soup = BeautifulSoup(response.text, 'html.parser')
table =soup.find_all('table')[0]
df = pd.read_html(str(table), decimal=',', thousands='.')[0]
pd.set_option('display.float_format', lambda x: f'{x:,.4f}')
df.head()
```

# Limpando dataset

## Convertendo string com % em float
* Div.Yield
* Mrg Ebit
* Mrg. Líq.
* ROIC
* ROE
* Cresc. Rec.5a

```python
df['Div.Yield'] = df['Div.Yield'].str.replace(',', '.')
df['Div.Yield'] = df['Div.Yield'].str.rstrip('%').astype('float') / 100.0

df['Mrg Ebit'] = df['Mrg Ebit'].str.replace('.', '')
df['Mrg Ebit'] = df['Mrg Ebit'].str.replace(',', '.')
df['Mrg Ebit'] = df['Mrg Ebit'].str.rstrip('%').astype('float') / 100.0

df['Mrg. Líq.'] = df['Mrg. Líq.'].str.replace('.', '')
df['Mrg. Líq.'] = df['Mrg. Líq.'].str.replace(',', '.')
df['Mrg. Líq.'] = df['Mrg. Líq.'].str.rstrip('%').astype('float') / 100.0

df['ROIC'] = df['ROIC'].str.replace('.', '')
df['ROIC'] = df['ROIC'].str.replace(',', '.')
df['ROIC'] = df['ROIC'].str.rstrip('%').astype('float') / 100.0

df['ROE'] = df['ROE'].str.replace('.', '')
df['ROE'] = df['ROE'].str.replace(',', '.')
df['ROE'] = df['ROE'].str.rstrip('%').astype('float') / 100.0


df['Cresc. Rec.5a'] = df['Cresc. Rec.5a'].str.replace('.', '')
df['Cresc. Rec.5a'] = df['Cresc. Rec.5a'].str.replace(',', '.')
df['Cresc. Rec.5a'] = df['Cresc. Rec.5a'].str.rstrip('%').astype('float') / 100.0


df.head(15)
```

# Applicando Filtros

```python
df_liquidez = df.loc[df['Liq.2meses'] > 1000000]
df_patrimonio = df_liquidez.loc[df['Patrim. Líq'] > 0]
df_ranking = df_patrimonio[['Papel', 'Cotação', 'ROIC', 'EV/EBIT']]

df_ev_sorted = df_ranking.sort_values('EV/EBIT')
df_ev_sorted = df_ev_sorted.reset_index(drop=True)
df_ev_sorted['ev/ebit-ranking'] = df_ev_sorted.index

df_roic_sorted = df_ev_sorted.sort_values('ROIC', ascending=False)
df_roic_sorted = df_roic_sorted.reset_index(drop=True)
df_roic_sorted['roic-ranking'] = df_roic_sorted.index

df_ranking = df_roic_sorted

df_ranking['score'] = df_ranking['ev/ebit-ranking'] + df_ranking['roic-ranking']
df_ranking = df_ranking.sort_values('score')
df_ranking = df_ranking.reset_index(drop=True)

df_ranking['Papel'] = df_ranking['Papel'] + '.SA'

df_ranking.head(60)
```

## Capturando informações adicionais

```python
def getSector(ticker):
    info = yf.Ticker(ticker)
    return info.info['sector']

def getShortName(ticker):
    info = yf.Ticker(ticker)
    return info.info['shortName']

getSector('B3SA3.SA')
```

```python
tickers = df_ranking['Papel']
sectorDict = {}

for ticker in tickers:
    try:
        print('getting sector in: ', ticker)
        sectorDict[ticker] = getSector(ticker)
    except:
        print('Error: ', ticker)
        sectorDict[ticker] = 'nan'

df_stockSector = pd.DataFrame.from_dict(sectorDict, orient='index').reset_index()
df_stockSector.columns = ['Papel', 'Setor']
df_stockSector
```

```python
df_ranking = pd.merge(df_ranking, df_stockSector, on='Papel', how='inner')
df_ranking
```

```python
df_ranking['Setor'].unique()
```

```python
df_ranking = df_ranking[df_ranking['Setor'] != 'Financial Services']
df_ranking = df_ranking[df_ranking['Setor'] != 'nan']
df_ranking.head(30)
```

```python
nameDict = {}

for ticker in tickers:
    try:
        print('getting shortName in: ', ticker)
        nameDict[ticker] = getShortName(ticker)
    except:
        print('Error: ', ticker)
        nameDict[ticker] = 'nan'

df_stockName = pd.DataFrame.from_dict(nameDict, orient='index').reset_index()
df_stockName.columns = ['Papel', 'Setor']
df_stockName
```
